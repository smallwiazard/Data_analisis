```python
# neural nets
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import pandas as pd
import numpy as np
# normalization and test sample
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading and prepearing dataset
apple_df = pd.read_csv("/kaggle/input/apple-revenue-from-1980-to-2022/applerevenue .csv")
apple_df = apple_df.drop(["date"], axis=1)
X = apple_df.drop(["profit or not"], axis=1)
y = apple_df["profit or not"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(X_train.shape)
print(y_train.shape)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
# model
neural_net = keras.Sequential()
neural_net.add(layers.Dense(32, activation="relu"))
neural_net.add(layers.Dense(32, activation="relu"))
neural_net.add(layers.Dense(32, activation="relu"))
neural_net.add(layers.Flatten())
neural_net.add(layers.Dense(1, activation="sigmoid"))
neural_net.build(input_shape=(None, 5)) 
```

```python
# compiling and fitting neural net
my_callbacks = [tf.keras.callbacks.EarlyStopping(patience=2)]
neural_net.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
history = neural_net.fit(
                    X_train,
                    y_train,
                    epochs=10,
                    batch_size=64,
                    callbacks = my_callbacks,
                    validation_split=0.2, verbose = 2)
```

```python
score = neural_net.evaluate(X_test, y_test, verbose=0) 
print('Test score:', score[0]) 
print('Test accuracy:', score[1])
```
